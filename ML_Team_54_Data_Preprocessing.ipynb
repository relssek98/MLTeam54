####Import Data from Database

```python
!pip install nltk

import csv
import nltk
from nltk.stem.porter import PorterStemmer
import numpy as np
import pandas as pd

text = []  
# Open file, must be uploaded to colab for every runtime
#with open('/content/drive/MyDrive/Machine Learning/Data/Suicide_Detection.csv') as file_obj:
with open('/content/drive/MyDrive/Machine Learning Project/Data/Suicide_Detection.csv') as file_obj:
            
    # Create reader object by passing the file 
    # object to reader method

    #potential lowerca
    #lower_stream = (line.lower() for line in file_obj)
    #csv_reader = csv.reader(lower_stream, delimiter=' ', quotechar='"', escapechar='^')
    
    reader_obj = csv.reader(file_obj)
    
    
    # Iterate over each row in the csv 
    #file using reader object
    for row in reader_obj:
      if row[0] is not '':
        if int(row[0]) < 250000:
          text.append(row)
data = {"raw": [], "label": []}
for i in range(len(text)):
  data["raw"].append(text[i][1])
  data["label"].append(text[i][2])
data = pd.DataFrame(data)
```

####Remove punctuation

```python
#library that contains punctuation
import string
string.punctuation

#defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
data["no_punctuation"]= data["raw"].apply(lambda x:remove_punctuation(x))
data.head()
```

####Convert Contractions/Abbreviations/Acronyms into words 


```python
!pip install contractions
import contractions
def remove_contractions(text):
  # creating an empty list
  expanded_words = []   
  for word in ''.join(text).split():
    # using contractions.fix to expand the shortened words
    expanded_words.append(contractions.fix(word))  

  expanded_text = ' '.join(expanded_words)
  return expanded_text
data['no_contractions']= data['no_punctuation'].apply(lambda x:remove_contractions(x))
data.head()
```

####Remove Emojis and URLs

```python
#defining the function to remove emoji
!pip install clean-text
from cleantext import clean
import re
def remove_emojiAndURL(text):
    emojiFree= clean(text, no_emoji=True)
    emojiFree = re.sub(r'http\S+', '', emojiFree)
    #print(emojiFree)
    return emojiFree
#storing the emoji free text
data["no_emojisURLs"]= data["no_contractions"].apply(lambda x:remove_emojiAndURL(x))
data.head()

```

####Tokenization

```python
#defining function for tokenization
from nltk.tokenize import word_tokenize
nltk.download('punkt')

data['tokenized']= data['no_emojisURLs'].apply(word_tokenize)
data.head()
```

####Stemming

```python
#defining the object for stemming
porter_stemmer = PorterStemmer()
#defining a function for stemming
def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text
data['stemmed']=data['tokenized'].apply(lambda x: stemming(x))
data.head()
```

####Stop Words

```python
#Stop words present in the library
#stopwords = nltk.corpus.stopwords.words('english')
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    output= [i for i in text if i not in stop]
    return output
#applying the function
data['no_stopwords']= data['stemmed'].apply(lambda x:remove_stopwords(x))
data.head()
```

####Lowercase Text 

```python
def lower(text):
  lower = [word.lower() for word in text]
  return lower

#storing the URL free text
data["clean"]= data["no_stopwords"].apply(lambda x:lower(x))
data.head()

```
